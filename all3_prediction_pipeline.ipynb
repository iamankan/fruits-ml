```python
from pathlib import Path
from PIL import Image
import numpy as np
from skimage import transform
import keras
```

```python
data = Path('all3')
```

```python
quality_modelpath = Path('all3_data_quality_model/')
bad_modelpath = Path('all3_data_type_model_bad/')
good_modelpath = Path('all3_data_type_model_good/')
mixed_modelpath = Path('all3_data_type_model_mixed/')
```

```python
quality_model = keras.models.load_model(quality_modelpath)
bad_model = keras.models.load_model(bad_modelpath)
good_model = keras.models.load_model(good_modelpath)
mixed_model = keras.models.load_model(mixed_modelpath)
```

```python
def load(filename: str, w: int or float, h: int or float, ch: int or float) -> np.ndarray:
   np_image = Image.open(filename)
   np_image = np.array(np_image).astype('float32')
   np_image = transform.resize(np_image, (w, h, ch))
   np_image = np.expand_dims(np_image, axis=0)
   return np_image
```

```python


bad = {0: 'Apple',
       1: 'Banana',
       2: 'Guava',
       3: 'Lime' ,
       4: 'Orange',
       5: 'Pomegranate'
      }

good = {0: 'Apple',
       1: 'Banana',
       2: 'Guava',
       3: 'Lime' ,
       4: 'Orange',
       5: 'Pomegranate'
      }

mixed = {0: 'Apple',
       1: 'Banana',
       2: 'Guava',
       3: 'Lime' ,
       4: 'Orange',
       5: 'Pomegranate'
      }

quality = {0:{name:'Bad', model:bad_model, ref: bad}, 
           1:{name:'Good', model:good_model, ref: good}, 
           2:{name:'Mixed', model:mixed_model, ref: mixed}
          }
```

```python
hit=0
miss=0
for fruits in data.iterdir():
    fruitsplit = fruits.split('_')
    fruitname = fruitsplit[0]
    fruitquality = fruitsplit[1]
    for samples in fruits.iterdir():
        image = load(samples, 224, 224, 3)
        qualidx = np.argmax(quality_model.predict(image))
        fruitidx = np.argmax(quality[qualidx]['model'].predict(image))
        if quality[qual]['ref'][fruitidx].lower() == fruitname.lower() and quality[qualidx]['name'].lower() == fruitquality.lower():
            hit = hit+1
        else:
            miss = miss+1
```
