```python
from PIL import Image
import numpy as np
from skimage import transform
```

```python
data_folder = 'all_data'
```

```python
file1 = 'all_data/Apple_Bad/IMG20200728175856.jpg'
```

```python
def load(filename: str, w: int or float, h: int or float, ch: int or float) -> np.ndarray:
   np_image = Image.open(filename)
   np_image = np.array(np_image).astype('float32')
   np_image = transform.resize(np_image, (w, h, ch))
   np_image = np.expand_dims(np_image, axis=0)
   return np_image

def get_index(name: str, my_list: list) -> int:
    return my_list.index(name)
```

```python
fruit_idx = ['Apple', 'Banana', 'Guava', 'Lime', 'Orange', 'Pomegranate']
fruit_quality = ['Bad', 'Good']
```

```python
image = load(file1, 224, 224, 3)
```

```python
np.argmax(type_model.predict(image))
```

```python
round(quality_model.predict(image)[0][0])
```

```python
from pathlib import Path
```

```python
hit = 0
miss = 0
p = Path('all_data')
for label in p.iterdir():
    print(f'NAME: {label.name}')
    plabel = Path(label)
    fruit = get_index(label.name.split('_')[0], fruit_idx)
    quality = get_index(label.name.split('_')[1], fruit_quality)
    for sample in plabel.iterdir():
        image = load(sample, 224, 224, 3)
        if fruit == np.argmax(type_model.predict(image)) and quality == round(quality_model.predict(image)[0][0]):
            hit = hit + 1
        else:
            miss = miss + 1
```
