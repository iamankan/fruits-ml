```python
good = fname+'/'+'Good_Quality_Fruits'
bad = fname + '/' + 'Bad_Quality_Fruits'
```

```python
goodclass = os.listdir(good)
```

```python
for lbl in goodclass:
    fruit = lbl.split('_')[0]
    samples = os.listdir(good+'/'+lbl)
    makedirectory(newdata+'/'+fruit)
    print('{} : {}'.format(fruit, len(samples)))
    shutil.copytree(good+'/'+lbl, newdata+'/'+fruit)
```

```python
badclass = os.listdir(bad)
```

```python
from distutils.dir_util import copy_tree
for lbl in badclass:
    fruit = lbl.split('_')[0]
    samples = os.listdir(bad+'/'+lbl)
    makedirectory(newdata+'/'+fruit)
    print('{} : {}'.format(fruit, len(samples)))
    copy_tree(bad+'/'+lbl, newdata+'/'+fruit)
```

## Split the data

```python
datapath = 'fruit_data'
trainpath = '/'.join([datapath,'train'])
testpath = '/'.join([datapath,'test'])
```

```python
import math
import shutil
import os
```

```python
trainfruits = os.listdir(trainpath)
for trainfruit in trainfruits:
    trainsamples = os.listdir('/'.join([trainpath, trainfruit]))
    n = len(trainsamples)
    testlength = math.floor(0.3*n)
    print(testlength)
    for trainsample in trainsamples:
        srcfile = '/'.join([trainpath, trainfruit, trainsample])
        destfile = '/'.join([testpath, trainfruit, trainsample])
#         print('{} -> {}'.format(srcfile, destfile))
        shutil.move(srcfile, destfile)
        testlength = testlength - 1
        print('testlength: {}'.format(testlength))
        if testlength == 0:
            break
```

```python
trainfruits = os.listdir(trainpath)
for trainfruit in trainfruits:
    if not os.path.exists('/'.join([testpath, trainfruit])):
        os.mkdir('/'.join([testpath, trainfruit]))
    trainsamples = os.listdir('/'.join([trainpath, trainfruit]))
    n = len(trainsamples)
    print(n)
```
